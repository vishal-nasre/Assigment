2. Write a Python program to reverse a word after accepting the input from the user.

Sample Output:

Input word: ineuron

Output: norueni

```python
word = input("Enter word to reverse it: ").split(" ")

for i in word:
    reverse_word = i[::-1]
    new_word = "".join(reverse_word)
    
print("\nThe Original Word = ", word)
print("The Reversed Word = ", new_word)
```
